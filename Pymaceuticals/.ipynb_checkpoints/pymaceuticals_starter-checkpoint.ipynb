```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merge_df = pd.merge(study_results, mouse_metadata, how='outer', on='Mouse ID')
# Display the data table for preview
merge_df
```

```python
# Checking the number of mice.
merge_df['Mouse ID'].value_counts().shape[0]
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dupe = merge_df.duplicated(subset=['Mouse ID','Timepoint'])
duplicate_rows = merge_df.loc[dupe]
duplicate_ids = duplicate_rows['Mouse ID'].unique()
duplicate_ids
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_rows_sorted = duplicate_rows.reset_index(drop=True).sort_values(by=['Mouse ID','Timepoint'])
duplicate_rows_sorted
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
merge_df[merge_df['Mouse ID'] == 'g989']
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = merge_df[merge_df['Mouse ID'] != 'g989']
clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_df['Mouse ID'].nunique()
```

## Quartiles, Outliers and Boxplots

```python
# clean_df['Drug Regimen'].nunique()

Final_Tumor_Size = clean_df.drop_duplicates(subset = ("Mouse ID"), keep = "last")

Final_Tumor_Size

Final_Tumor_by_drug = Final_Tumor_Size[["Mouse ID", "Drug Regimen", "Tumor Volume (mm3)"]]


Final_Tumor_Size_Capomulin = Final_Tumor_by_drug.loc[Final_Tumor_by_drug["Drug Regimen"] == "Capomulin"]
Final_Tumor_Size_Ramicane = Final_Tumor_by_drug.loc[Final_Tumor_by_drug["Drug Regimen"] == "Ramicane"]
Final_Tumor_Size_Infubinol = Final_Tumor_by_drug.loc[Final_Tumor_by_drug["Drug Regimen"] == "Infubinol"]
Final_Tumor_Size_Ceftamin = Final_Tumor_by_drug.loc[Final_Tumor_by_drug["Drug Regimen"] == "Ceftamin"]
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_timepoints = clean_df.groupby('Mouse ID')['Timepoint'].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

final_tumor_volume_df = pd.merge(clean_df, max_timepoints, on=['Mouse ID', 'Timepoint'], how='inner')

Final_Tumor_by_drug = final_tumor_volume_df[["Mouse ID", "Drug Regimen", "Tumor Volume (mm3)"]]


Final_Tumor_Size_Capomulin = Final_Tumor_by_drug.loc[Final_Tumor_by_drug["Drug Regimen"] == "Capomulin"]
Final_Tumor_Size_Ramicane = Final_Tumor_by_drug.loc[Final_Tumor_by_drug["Drug Regimen"] == "Ramicane"]
Final_Tumor_Size_Infubinol = Final_Tumor_by_drug.loc[Final_Tumor_by_drug["Drug Regimen"] == "Infubinol"]
Final_Tumor_Size_Ceftamin = Final_Tumor_by_drug.loc[Final_Tumor_by_drug["Drug Regimen"] == "Ceftamin"]


# final_tumor_volume_df = final_tumor_volume_df[final_tumor_volume_df['Drug Regimen'].isin(regimens)]

final_tumor_volume_df
```

```python
box_plot = pd.DataFrame({"Capomulin" : Final_Tumor_Size_Capomulin["Tumor Volume (mm3)"],
                            "Ramicane" : Final_Tumor_Size_Ramicane["Tumor Volume (mm3)"],
                           "Infubinol" : Final_Tumor_Size_Infubinol["Tumor Volume (mm3)"],
                        "Ceftamin" : Final_Tumor_Size_Ceftamin["Tumor Volume (mm3)"]})

box_plot
```

```python
avg_tumor_size_final = round(box_plot.mean(),2)
quartiles = round(box_plot.quantile([.25,.5,.75]),2)
lower_q = quartiles.loc[.25]
upper_q = quartiles.loc[.75]
iqr = upper_q - lower_q
lower_bound = lower_q - (1.5*iqr)
upper_bound = upper_q + (1.5*iqr)
outlier_df = pd.DataFrame({"Lower Bound" : lower_bound, "Upper Bound" : upper_bound})

print(f"Average Final Tumor Size\n\n{avg_tumor_size_final}\n\n")

print(f"Tumor Size Quartiles\n\n{quartiles}\n\n")

print(f"Anything outside of these ranges is considered an outlier\n\n{outlier_df}\n\n")

Capomulin_outlier = Final_Tumor_Size_Capomulin.loc[(Final_Tumor_Size_Capomulin["Tumor Volume (mm3)"] <= outlier_df.loc["Capomulin", "Lower Bound"]) |
                                                   (Final_Tumor_Size_Capomulin["Tumor Volume (mm3)"] >= outlier_df.loc["Capomulin", "Upper Bound"])]
Ramicane_outlier = Final_Tumor_Size_Ramicane.loc[(Final_Tumor_Size_Ramicane["Tumor Volume (mm3)"] <= outlier_df.loc["Ramicane", "Lower Bound"]) |
                                                   (Final_Tumor_Size_Ramicane["Tumor Volume (mm3)"] >= outlier_df.loc["Ramicane", "Upper Bound"])]
Infubinol_outlier = Final_Tumor_Size_Infubinol.loc[(Final_Tumor_Size_Infubinol["Tumor Volume (mm3)"] <= outlier_df.loc["Infubinol", "Lower Bound"]) |
                                                   (Final_Tumor_Size_Infubinol["Tumor Volume (mm3)"] >= outlier_df.loc["Infubinol", "Upper Bound"])]
Ceftamin_outlier = Final_Tumor_Size_Ceftamin.loc[(Final_Tumor_Size_Ceftamin["Tumor Volume (mm3)"] <= outlier_df.loc["Ceftamin", "Lower Bound"]) |
                                                   (Final_Tumor_Size_Ceftamin["Tumor Volume (mm3)"] >= outlier_df.loc["Ceftamin", "Upper Bound"])]

outlier_list = pd.concat([Capomulin_outlier, Ramicane_outlier, Infubinol_outlier, Ceftamin_outlier])

print(f"Below is a list of all outliers\n\n{outlier_list}")
```

```python
# Put treatments into a list for the for loop and plot labels
regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create an empty list to store potential outliers
outliers = {}

# Iterate through each regimen
for regimen in regimens:
    # Filter the DataFrame to get data for the current regimen
    tumor_volume_data = box_plot[regimen]

    
    # Calculate quartiles and IQR
    quartiles = tumor_volume_data.quantile([0.25, 0.75])
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    iqr = upper_q - lower_q
    lower_bound = lower_q - (1.5 * iqr)
    upper_bound = upper_q + (1.5 * iqr)
    
    # Identify potential outliers
    potential_outliers = tumor_volume_data[(tumor_volume_data < lower_bound) | (tumor_volume_data > upper_bound)]
    
    # Append potential outliers to the list
    outliers[regimen] = potential_outliers

# Print potential outliers for each regimen
for regimen, potential_outliers in outliers.items():
    print(f"{regimen}'s potential outliers:\n{potential_outliers}\n")
```

```python





capomulin_averages = clean_df[clean_df["Drug Regimen"] == "Capomulin"].groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
mouse_weights = clean_df[clean_df["Drug Regimen"] == "Capomulin"].groupby("Mouse ID")["Weight (g)"].first()


plt.scatter(capomulin_averages, mouse_weights, facecolor = "blue")
plt.title("Average Mouse Weight vs. Average Tumor Volume")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

slope, y_int, r_value, p_value, std_err = st.linregress(capomulin_averages, mouse_weights)

fit = capomulin_averages * slope  + y_int
equation = "y= " + str(round(slope,2)) + "x + " + str(round(y_int,2))
r_square = "R-Squared Value is " + str(round(r_value,2))
plt.scatter(capomulin_averages, mouse_weights, facecolor = "blue")
plt.plot(capomulin_averages, fit, color = "red")
plt.annotate(equation, (20, 40), color = "red", fontsize = 12)
plt.annotate(r_square, (19, 38.5), color = "red", fontsize = 12)
plt.title("Average Mouse Weight vs. Average Tumor Volume")
plt.xlabel("Weight(g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
